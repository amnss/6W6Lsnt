```python
import numpy as np

# Function for TDMA
def tdma(l, u, d, b):
    n = len(d)
    x = np.zeros(n)
    for i in range(1, n):
        factor = l[i] / d[i - 1]
        d[i] -= factor * u[i - 1]
        b[i] -= factor * b[i - 1]
    x[-1] = b[-1] / d[-1]
    for i in range(n - 2, -1, -1):
        x[i] = (b[i] - u[i] * x[i + 1]) / d[i]
    return x

# Checking if a matrix is diagonally dominant
def is_diagonally_dominant(matrix):
    diag = np.abs(matrix.diagonal())
    sums = np.sum(np.abs(matrix), axis=1) - diag
    return np.all(diag >= sums)

# Input method choice
print("Choose the method:")
print("1. TDMA")
print("2. Gauss-Seidel")
print("3. Newton-Raphson")
method = int(input("Enter the number corresponding to the method: "))

if method == 1:
    n = int(input("Enter the number of equations: "))
    l=np.ones(n)
    l[n-1]=0
    u=np.ones(n)
    u[0]=0
    u[n-1]=0
    d=np.ones(n)
    for i in range(1,n-1):
        d[i]=-2
    b=np.zeros(n)
    b[0]=100
    b[n-1]=25
    x = tdma(l, u, d, b)
    print("Solution x:")
    print(np.array(x**0.5))

elif method == 2:
    n = int(input("Enter the number of equations: "))
    A = np.zeros((n, n))
    # print("Enter matrix A (seperated by space):")
    # for i in range(n):
    #     A[i] = list(map(float, input().split()))
    A = np.zeros((n, n))
    A[0,0]=1
    for i in range(1, n - 1):
        A[i, i - 1] = 1
        A[i, i] = -2
        A[i, i + 1] = 1
    A[n-1,n-1]=1
    # Input vector B
    # print("Enter vector B (seperated by space):")
    # B = np.array(list(map(float, input().split())))
    B=np.zeros(n)
    B[0]=100
    B[n-1]=25
    # Input initial guess vector G
    # print("Enter initial guess vector G (space-separated values):")
    G = np.ones(n)


    error_percent = 0.1

    # Calculate allowable error using the sum-norm error formula
    allowable_error = (error_percent / 100)
    # Input allowable error
    #eps = float(input("Enter error to stop: "))

    # Maximum number of iterations
    max_iterations = 1000

    # Check if the matrix is diagonally dominant, if not, apply pivoting
    if not is_diagonally_dominant(A):
        print("After using pivoting.")
        for i in range(n):
            max_row = i
            for j in range(i + 1, n):
                if abs(A[j, i]) > abs(A[max_row, i]):
                    max_row = j
            # Swapping rows
            A[[i, max_row]] = A[[max_row, i]]
            B[[i, max_row]] = B[[max_row, i]]

    # Gauss-Seidel method
    for _ in range(max_iterations):
       # print(G)
        G_old = np.copy(G)
        for i in range(n):
            sigma = 0
            for j in range(n):
                if j != i:
                    sigma += A[i, j] * G[j]
            G[i] = (B[i] - sigma) / A[i, i]
        # Checking for convergence

        if np.linalg.norm(G - G_old, np.inf) < allowable_error:
            break

    # Output the solution
    if np.linalg.norm(G - G_old, np.inf) < allowable_error:
        print("Solution x:")
        print(G**0.5)
    else:
        print("Did not converge")

elif(method==3):
    def initialize_h(n):
        h = np.ones(n)
        return h

    def equations(h):
        n = len(h)
        f = np.zeros(n)
        f[0]=h[0]-10
        for i in range(1, n - 1):
            f[i] = h[i - 1]**2 - 2 * h[i]**2 + h[i + 1]**2
        f[n-1]=h[n-1]-5
        return f

    def jacobian(h):
        n = len(h)
        J = np.zeros((n, n))
        J[0,0]=1
        for i in range(1, n - 1):
            J[i, i - 1] = 2 * h[i - 1]
            J[i, i] = -4 * h[i]
            J[i, i + 1] = 2 * h[i + 1]
        J[n-1,n-1]=1
        return J

    def newton_raphson(n, max_iterations=100, tolerance=1e-4):
        h = initialize_h(n)
        for iteration in range(max_iterations):
            f = equations(h)
            J = jacobian(h)
            print(h)
            delta_h = np.dot(np.linalg.inv(J),-f)
            h+= delta_h
            if np.linalg.norm(delta_h) < tolerance:
                return h

        raise Exception("Newton-Raphson did not converge")

    n = int(input("Enter the number of equations (n): "))
    solution = newton_raphson(n)
    print("Solution:")
    print(solution)
else:
    print("Change the choice")
```
